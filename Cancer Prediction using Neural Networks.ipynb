```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(cancer_data.data, columns = cancer_data.feature_names)
```

```python
df["label"] = cancer_data.target
```

```python
X = df.drop('label', axis=1)
Y = df['label']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

# Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

# Building the Neural Network

Importing dependencies

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

1 Neuron for each input feature

No of neurons in input layer = no of features in dataset

1 Neuron for each output class

No of neurons in output layer = no of classes in output

```python
# Setting up the layers of Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),            # INPUT LAYER
    keras.layers.Dense(20, activation = 'relu'),        # HIDDEN LAYER, relu = Rectified Linear Unit
    keras.layers.Dense(2, activation = 'sigmoid')       # OUTPUT LAYER
])
```

```python
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',   # Used for label encoding and not one hot encoding.
              metrics = ['accuracy'])
```

Training the neural network

```python
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs = 15)

# Loss function should be MINIMUM -> difference between real and predicted value
```

Vizualising accuracy and loss

```python
plt.plot(history.history['accuracy'])

plt.title("Model accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
```

```python
plt.plot(history.history['loss'])

plt.title("Model loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
```

Accuracy of model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
```

```python
Y_pred = model.predict(X_test_std)
```

Converting the prediction probability to class labels

```python
Y_pred_labels = [np.argmax(i) for i in Y_pred]   #np.argmax gives index of maximum value
```
